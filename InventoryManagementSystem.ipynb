# Add Records to inventory

```python
def AddRecord() :
    prod_id = int(input("Enter the id of product : "))
    name = input("Enter the name of product")
    price = int(input("Enter the price : "))
    quant = int(input("Enter the quantity of product : "))
    ptype = input("Enter the product type : ")
    rating = int(input("Entert the rating : "))
    
    fd = open("inventory.json",'r')
    r = fd.read()
    fd.close()
    record = json.loads(r)
    
    record[prod_id] ={}
    record[prod_id]["name"] = name
    record[prod_id]["price"] =price
    record[prod_id]["quantity"] = price
    record[prod_id]["product_type"] = ptype
    record[prod_id]["Consumer_rating"] = rating
    
    print("Record has been added Successfully!!")
    
    f = open("inventory.json","w")
    jso = json.dumps(record) 
    f.write(jso)
    f.close()
    
```

# Buy Product from store

```python
def BuyProduct() :
    number = int(input("Enter the number of products to purchase"))
    
    fd = open("inventory.json","r")
    r = fd.read()
    fd.close()
    
    items = json.loads(r)
    
    bill = open("Generated_Bill.txt","w")
    sale = open("Sales_Analysis.txt","w")
    
    total =0
    _max =0

    
    for i in range(number) :
        _id = input("Enter the id of product : ")
        quant = int(input("Enter the quantity of product : "))
        if items[_id]["quantity"] > 0 :
            print("Item purchased : " , items[_id]["name"])
            print("Price : " , items[_id]["price"])
            print("Product Category : " ,items[_id]["product_type"])
            items[_id]["quantity"] = items[_id]["quantity"] - quant
            total = total+items[_id]["price"]*quant
            bill.write("Item purchased : {} \nPrice : {} \nProduct Category : {} \n Quantity : {} \n".format(str(items[_id]["name"]) , str(items[_id]["price"]),str(items[_id]["product_type"]),str(quant)))       
            if(quant>_max) :
                _max =quant
                prod = items[_id]["name"]
                val = items[_id]["price"]*quant
                cat = items[_id]["product_type"]
        else :
                print("Insufficient amount of product in inventory :(")
    
    print("Total Bill of yours is : " , total)
    print("Collect Bill\n")
    bill.write("\n\n\n\nTotal Bill : {}".format(str(total)))
    
    fd = open("inventory.json","w")
    js = json.dumps(items)
    fd.write(js)
    fd.close()
    
    bill.close()
        
    sale.write("Sales Analysis \nMaximum Items bought from store :\n Item name : {} \nQuantity : {} \nTotal Value : {} \nCategory : {} \n".format(str(prod),str(_max),str(val),str(cat)))
    sale.close()
```

# Update the items in Inventory

```python
def UpdateRecord() :
    _id = input("Enter the id of product to update : ")
    attr = input("Enter the attribute to change ")
    value = input("Enter the value of selected attribute")
    
    fd = open("inventory.json","r")
    r = fd.read()
    fd.close()
    
    record = json.loads(r)
    
    if(attr == "name") :
            record[_id]["name"] = value
    elif(attr =="price") :
            record[_id]["price"] = int(value)
    elif(attr =="quantity") :
            record[_id]["quantity"] = int(value)
    elif(attr == "product_type") :
            record[_id]["product_type"] = value
    else :
            record[_id]["Consumer_rating"] = int(value)
            
    print("Database Updated!!")

                
```

# Feedback/ Rating for Product

```python
def feedback() :
    _id = input("Enter the product id")
    ra = int(input("Enter the customer rating of the product"))
    
    fd = open("Number_User_rating.json","r")
    rec= fd.read()
    fd.close()
    
    record = json.loads(rec)
    users = int(record["total_user"])
    
    f = open("inventory.json","r")
    re = f.read()
    f.close()
    
    r = json.loads(re)
    
    record["total_user"] +=1
    new_rating = int((users*r[_id]["Consumer_rating"] + ra)/(users+1))
    
    r[_id]["Consumer_rating"] = new_rating
    
    f= open("inventory.json","w")
    js = json.dumps(r)
    f.write(js)
    f.close()
    
    fd = open("Number_User_rating.json","w")
    jso =json.dumps(record)
    fd.write(jso)
    fd.close()
    
```

# Recommendation System for User

```python
def Recommendation() :
    fd = open("inventory.json","r")
    r =fd.read()
    fd.close()
    
    record = json.loads(r)
    
    lst =[]
    records = dict(record)
    for key in records :
        if(record[key]["Consumer_rating"] >=8) :
            lst.append(record[key]["name"])
    
    print("Items Recommended for you : \n")
    print(lst)
```
